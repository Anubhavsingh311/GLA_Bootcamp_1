```python
students = [
    {"name": "Ansh", "grade": 85},
    {"name": "Anubhav", "grade": 60},
    {"name": "Didi", "grade": 99},
    {"name": "Hardik", "grade": 55},
    {"name": "Dev", "grade": 95}
]

for student in students:
    print("Student:", student["name"])
    print("Marks:", student["grade"])

    if student["grade"] >= 75:
        print("Performance: Strong")
    elif student["grade"] >= 50:
        print("Performance: Average")
    else:
        print("Performance: Needs Help")
```
